This cell prints the results of 100 rolls of a 20-sided dice.

```python
#import random
import numpy

AllDiceRolls = numpy.random.randint(1,20,100)
print(AllDiceRolls)
```

This cell finds the average of the 100 rolls.

```python
AverageRoll = numpy.mean(AllDiceRolls)
print('the average roll is:')
print(AverageRoll)
```

```python
STD = numpy.std(AllDiceRolls)
print('the standard deviation is:')
print(STD)
```
